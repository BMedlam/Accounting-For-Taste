# Capstone Workbook 4: Initial Modelling

This notebook contains the initial modelling for the processed dataframe.

Models techniques used in this workbook are: Linear Regression; Lasso (Linear Regression with a L1 penalty); and Random Forest.

The models will just be completed using the columns which numerical values, including those that have been converted to be represented numerically during the preprocessing stage. 

This initial modelling stage will give an insight into how effective the various models are in predicting the unseen target data, after being trained on a training data subset. Columns that are especially effective at predicting the target column will be identified and insights can be used to inform the advanced modelling stage.

```python
# Import libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
import statsmodels.api as sm
```

```python
# Import data 
airbnb_ldn = pd.read_csv('airbnb_ldn_pp.csv')
```

```python
# drop 'Unnamed: 0'
airbnb_ldn = airbnb_ldn.drop(columns = 'Unnamed: 0')
```

```python
airbnb_ldn.shape
```

For this inital modelling, a linear regression model will be used. This model will determine the qualities of the various numerical columns in determining the target variable 'Annual Revenue LTM'.

```python
num_cols = airbnb_ldn.select_dtypes(exclude='object')
```

```python
num_cols.info()
```

Now the dataframe has been imported and numerical variables separated, a few additional checks will be complete.

```python
# confirm whether there are any null values left in the dataframe:
num_cols.isnull().any().any()
```

Now it is confirmed that there are no null values in the dataframe. The different numerical columns will be evaluated. Those that do not serve any semantic value when predicting the target column will be dropped.

```python
# View numerical columns:
num_cols.head().T
```

The columns identified as being semantically irrelevant are: 
- Latitude
- Longitude
- Airbnb Host ID	
- Number of Bookings LTM - Number of observed month	

These columns will be dropped from the dataframe:

```python
num_cols.drop(columns = ['Latitude', 'Longitude', 'Airbnb Host ID', 'Number of Bookings LTM - Number of observed month'], inplace=True)
```

The dataframe will be split between the indepedent and target variables:

```python
X = num_cols.drop(columns = 'Annual Revenue LTM (Native)')
y = num_cols['Annual Revenue LTM (Native)']
```

The data will now be split between the train and test sets:

```python
# required imports 
from sklearn.model_selection import train_test_split
```

```python
# split independent and target data between train and test sets:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
# reset index for X train and test datasets
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
```

The indexes must be reset for the y train and test datasets. This involves having to convert the series data into a dataframe for the conversation to occur, with the dataframe then being returned to a series format:

```python
# reset index for y train and test datasets
# convert to dataframe
y_train = y_train.to_frame()
y_test = y_test.to_frame()

# reset index
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)

# return y datasets to series format
y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

Now the dataframe has been split into train and test variables, the values of the train data will be further investigated.

```python
X_train.shape
```

```python
X_train.head().T
```

The numerical variable features are made up of continuous and binary columns (many binary columns resulting from the previous one-hot-encoding of categorical columns). For effective modelling, the continuous variable must be scaled. These columns will be identified, separated, scaled and then concatenated back onto the original dataframe:

```python
# Isolate the binary columns
X_binary = X_train.columns[X_train.nunique() == 2]

# Create new dataframe by dropping the binary columns
X_cont = X_train.drop(columns = X_binary)
```

```python
X_cont
```

```python
X_cont.head()
```

As determined during the EDA phase, a log transformation for the continuous variables improves their correlation with the target variable and ensures the independent variables are more normally distributed.

```python
# log transform the inpdendent variables, adding 1 to prevent the log of zero being taken 
X_log = np.log(X_cont + 1)
```

Now the log for the independent variables has been found, these same variables will be scaled to improve the modelling effectiveness when comparing the values to the binary columns:

```python
# import scaling library
from sklearn.preprocessing import StandardScaler
```

```python
# scale the X_log train dataframe
X_train_ls = StandardScaler().fit_transform(X_log)
```

```python
# Convert the scaled array back into a dataframe
X_train_t = pd.DataFrame(X_train_ls, columns = X_cont.columns)

# View the dataframe
X_train_t
```

The log transformed and scaled continuous variables must be concatenated back together with the binary variables.

```python
# Create dataframe isolating the binary columns
X_binary_df = X_train[X_binary]

# Concatenate the continous and binary dataframes together:
X_train_ls = pd.concat([X_train_t, X_binary_df], axis=1)
```

```python
X_train_ls
```

## Model 1: Linear Regression - X values log transformed

An initial linear regression model will be made. A cross validation will be used to instantly test the effectiveness of the model. 

```python
# importing the required packages 
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 1. Instantiate the Model
linreg = LinearRegression()

# 2. Fit the model using 5 folds 
scores_m1 = cross_val_score(linreg, X_train_ls, y_train, cv=5)
print(scores_m1)
```

```python
# determine the average score from the 5 folds 
print(f"The average R-squared score from the first cross-validation model is {round(scores_m1.mean(),2)}.")
```

An initial cross validated linear regression accuracy average of 0.6 is a relatively good result. This has been completed using the log transformed continuous independent variables. Additional models will be completed with the standard independent values.

## Model 2: Linear Regression Model - Standard X values

Not using a log transformation on the independent features.

```python
# scale the standard x values:
scaler = StandardScaler()
X_t_s = scaler.fit_transform(X_cont)
```

```python
# Convert the scaled array back into a dataframe
X_t_s_df = pd.DataFrame(X_t_s, columns = X_cont.columns)

# View the dataframe
X_t_s_df
```

```python
# Concatenate the continous and binary dataframes together:
X_train_s = pd.concat([X_t_s_df, X_binary_df], axis=1)
```

```python
X_train_s
```

```python
# 2. Fit the model using 5 folds 
scores_m2 = cross_val_score(linreg, X_train_s, y_train, cv=5)
print(scores_m2)
```

```python
print(f"The average R-squared score from the second cross-validation model is {round(scores_m2.mean(),2)}.")
```

The model with the unlogged independent variables has performed better, with a R-squared value of 0.69; indicating that the 69% of the variance in the target column can be explained by the independent variables.

## Model 3: Lasso Model

Now that a successful initial linear regression model has been produced, a linear regression model using a L1 penalty (Lasso) will be complete. 

The purpose of this model is to aid in feature selection. The L1 penalty will alter the coefficients, and essentially bring the coefficients of non-influencial columns to zero; these columns can thus be dropped.

A GridSearchCV will be used, to determine which hyperparameters are optimal for determining the highest model accuracy.

The same scaled X_train data used in model 2 can be used for this model:

```python
# import lasso model
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
# define model type:
lasso = Lasso()

# define parameters:
param_grid = {'alpha': [0.1, 0.5, 1, 2],
            'max_iter' : [1000, 5000]}

# define the grid search:
Lasso_gs = GridSearchCV(estimator=lasso, param_grid=param_grid, cv=5, n_jobs=-1, verbose=5)

# grid search
Lasso_gs.fit(X_train_s, y_train)
```

```python
# determine the optimal parameters found within the gridsearch:
best_params = Lasso_gs.best_params_
best_params
```

```python
# determine the best score from the model:
print(f"The best accuracy score from the Lasso model is {round(Lasso_gs.best_score_, 2)}.")
```

```python
# make model using best parameters:
best_model = Lasso(**best_params)
```

```python
# fit model to train data:
best_model.fit(X_train_s, y_train)
```

```python
# Use fitted model to determine feature coefficients:
coefficients = best_model.coef_

# extract feature names 
feature_names = X_train_s.columns

coef_L1 = pd.DataFrame({'Features' : feature_names, 'Coefficients' : coefficients})
coef_L1
```

The table above, shows the coefficient values that have resulted from the lasso (L1 penalty) linear regression model. These coefficients will be plotted on a bar chart, so that the most influencial columns can be more easily identified, as well as those with zero coefficients.

```python
# create dataframe with the sorted coefficient values 
sort_coef = coef_L1.sort_values(by='Coefficients')

# plt the sorted values on a bar chart
plt.figure(figsize=(8, 8))
plt.barh(sort_coef['Features'], sort_coef['Coefficients'])
plt.xlabel('Lasso Model Coefficient Results')
plt.title('The Lasso Model Coefficients for the Numeric Independent Variables')
plt.show()
```

```python
# returning the columns with zero coefficient:
coef_L1[coef_L1['Coefficients'] == 0]
```

There are several columns that have coefficients very close to zero, looking to essentially be zero when displayed on the coefficient visualisation. When looking at the columns that have an actual value of zero, 'Cancellation Policy strict', 'Checkout Time_late' and 'Checkout Time_none' are the columns identified as having no influence on predicting the target column.

The columns with the highest absolute coefficients, which are the most influencial in determing the target variable can also be identified. Unsurprisingly, the 'Average Daily Rate (Native)' and 'Count Reservation Days LTM' have the two highest positive coefficients. It would make sense, that a property with a higher daily rate, that is occupied for a higher portion of the year, would result in a higher annual revenue per year. 

Other columns coefficients which provide more valuable insights can also be found. Looking at the positive coefficients, having a 'super strict' cancellation policy looks to be influencial in increasing yearly revenue. There are two explanations for this, the first, a very strict policy will essentially reduce the quantity of cancellations, meaning the occurance of last minute cancellations and loss of tenants and thus income will be reduced. But also, when considering the types of properties, it could be suggested the more affluent, higher quality properties, will naturally have a more strict cancellation policy. A property that is exclusive, expensive and in high demand, would be more likely to also have a stricter cancellation policy, with these various characteristics resulting in a higher annual revenue. The question therefore, is whether a stricter cancellation policy influences the yearly income, or whether the type of property that has high income due to its quality, is also more likely to have a super strict cancellation policy.

The coefficients for the continuous variables are the coefficients resulting in these variables being modelled using their 'StandardScaled' values. These coefficients will be inverted, so that they can be represented to show their true proportional impact on the target column.    

```python
X_cont.columns
```

```python
# isolate the continuous and binary column coefficients:
cont_coef = coef_L1[coef_L1['Features'].isin(X_cont.columns)]
binary_coef = coef_L1[~coef_L1['Features'].isin(X_cont.columns)]
```

```python
# inverse the continuous variables coefficients:
cont_coef['Coefficients'] = pd.DataFrame(scaler.inverse_transform(cont_coef['Coefficients']).reshape(-1, 1))
```

```python
# concatenate these continuous coefficients with the binary ones:
coef_L1 = pd.concat([cont_coef, binary_coef], axis=0)
```

```python
# View the newly created dataframe:
coef_L1
```

The above table shows the numeric values of the unscaled continuous variables, along with the original values for the binary column coefficients.

These coefficients will be visualised:

```python
# create dataframe with the sorted coefficient values 
sort_coef = coef_L1.sort_values(by='Coefficients')

# plt the sorted values on a bar chart
plt.figure(figsize=(8, 8))
plt.barh(sort_coef['Features'], sort_coef['Coefficients'])
plt.xlabel('Lasso Model Coefficient Results')
plt.title('The Lasso Model Coefficients for the Numeric Independent Variables')
plt.show()
```

The graph above shows how the unscaled variables now proportionally have a much larger resultant impact on the target column, when compared to the previous visual. The coefficient of the binary values have remained constant, but their relative size when visualised now looks significantly smaller when compared to the inverted continuous columns.

## Model 4: Decision Tree

A decision tree pipeline model will be used for the next modelling attempt. A grid search cross validation will be used, to optimise the various input parameters within the decision tree. 

```python
# Additional required imports 

# To "pickle" things, like accuracies or even an entire fitted model
import joblib

# To set up a temporary directory for caching pipeline results
from tempfile import mkdtemp

# To build a pipeline
from sklearn.pipeline import Pipeline

# import decision tree regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# To do a cross-validated grid search
from sklearn.model_selection import GridSearchCV
```

A pipeline will be created for the first decision tree model:

```python
# Set up a directory to cache the pipeline results
cachedir = mkdtemp()

# Set up a pipeline
# The steps here act as placeholders and will be changed when we pass the pipeline into the grid search later
estimators = [('dt', DecisionTreeRegressor(criterion='mse'))]
pipe = Pipeline(estimators, memory = cachedir)
```

```python
# params for DecisionTreeRegressor() : model 1
# max depth

max_depth = [6, 8, 10, 12]

# min samples required for split
min_samples_split = [6, 8, 10, 12]

# min 
min_leaf = [1, 2, 3, 4, 6]

max_features = [2, 4, 6, 8]

# define the parameter grid
params = dict(
    dt__min_samples_split = min_samples_split,
    dt__min_samples_leaf = min_leaf,
    dt__max_depth = max_depth)
```

```python
dt_gsCV = GridSearchCV(pipe, param_grid=params, cv=5, n_jobs=-1, verbose=5)
dt_gsCV.fit(X_train, y_train)
```

```python
print(f"The decision tree model score on the entire train dataset is:{round(dt_gsCV.score(X_train, y_train),3)}.")
print(f"The decision tree model score based on the grid search cross-validation is {round(dt_gsCV.best_score_,3)}.")
```

```python
# determine the best parameters for the decision tree model:
best_model = dt_gsCV.best_estimator_.get_params()['dt']
best_model
```

```python
# fit the best model to the training dataset
best_model.fit(X_train, y_train)

# score with the testing dataset
round(best_model.score(X_test, y_test), 3)
```

```python
# Make predictions
y_pred_train = best_model.predict(X_train)
y_pred = best_model.predict(X_test)



# Evaluate model
print(f'Train MAE --- {round(mean_absolute_error(y_train, y_pred_train),2)}')
print(f'Test MAE ---- {round(mean_absolute_error(y_test, y_pred),2)}')
print(f'Train R2 ---- {round(best_model.score(X_train, y_train),2)}')
print(f'Test R2 ----- {round(r2_score(y_test, y_pred),2)}')
```

Looking at the results from the decision tree model. The model has scored an R-squared value of 0.99 when fitting to the training dataset and 0.98 when fitting to the test data. This indicates very high accuracy for both the test and training data. 

The MAE scores are somewhat adequate, a score of 789 indicates the model prediction based on the training data for a specific point is 789 off the actual value of annual revenue. As the actual values of revenue vary significantly and go into the 100,000s, an absolute error of 789 is not too significant. However, the 'Test MAE' is larger, with a value of 1297. This isn't too significant in terms of the actual annual revenue values, but does show some difference to the training data MAE. This could indicate a small degree of overfitting, or just reflect the slight differences between the train and test datasets. 

For now this is an positive first decision tree model output. A more advanced decision tree model will be complete in the advanced modelling stage.
